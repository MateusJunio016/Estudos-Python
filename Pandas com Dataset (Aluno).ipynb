### 1 - Explorando o dataset:

```python
# Importando o Numpy e Pandas:
import pandas as pd
import numpy as np
```

```python
# Carregando arquivo .csv:
df = pd.read_csv('/content/supermarket.csv')
```

```python
# Mostrado os 05 primeiros registros do Data Frame:
df.head(2)
```

```python
# Mostrando os 05 últimos registros do Data Frame:
df.tail(2)
```

```python
# Visualizando os detalhes e dados estatísticos do Data Frame:
df.describe()
```

### 2 - Exportando para outros formatos:

### 3 - Selecionando os dados e realizando operações:

```python
# Recuperando os dados de uma coluna apenas:
df['City']
```

```python
# Recuperando dados de um grupo de colunas:
df[['City', 'Total']]
```

```python
# Aplicando o describe() apenas para uma coluna:
df['City'].describe()
```

```python
# Visualizado um resumo geral do Data Frame:
df.info()
```

```python
# Visualizando a quantidade de linhas x colunas do Data Frame:
df.shape
```

```python
# Ordenando os dados do Data Frame de acordo com uma das colunas:
df.sort_values('Total')
```

```python
# Visualizado apenas uma das colunas com os valores ordenados (ordem decrescente):
df['City'].sort_values(ascending=False)
```

```python
# Aplicando o filtro por determinado tipo de dado existente em determinada coluna:
df[
    df['Product line'] == 'Electronic accessories'
    ]
```

```python
# Aplicando o filtro com mais de uma condição:
df_test = df[
    (df['City'] == 'Yangon') &
    (df['Product line'] == 'Electronic accessories')
]

df_test.head(5)
```

```python
# Criando uma nova coluna e aplicando uma operação matemática entre as variáveis:
df['Ticket'] = df['Total'] / df['Quantity']
df.head(2)
```

### 4 - Utilizando funções agregadas:

```python
# Obtendo a média de valores de uma coluna:
df['Total'].mean()
```

```python
# Obtendo o valor mínimo de uma coluna:
df['Total'].min()
```

```python
# Obtendo o valor máximo de uma coluna:
df['Total'].max()
```

```python
# Retornando valores no quantil* fornecido sobre o eixo solicitado:
df.quantile(0.25)
```

```python
# Obtendo a moda de alguma coluna (valor que mais aparece):
df['Quantity'].mode()
```

```python
# Agregando mais de uma operação:
df['Rating'].agg([np.mean, max, min])
```

```python
df.groupby(['Payment']).Payment.count().sort_values().plot(kind="bar")
```
